```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x, y)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('title')
ax.set_xlim(x.min(), x.max())
ax.set_ylim(y.min(), y.max())
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
elements = [axes1, axes2]
for n in elements:
    n.plot(x, y, 'r')
    n.set_xlabel('X')
    n.set_ylabel('Y')
    n.set_xlim([x.min(), x.max()])
    n.set_ylim([y.min(), y.max()])
```

```python
fig2 = plt.figure()

axes3 = fig2.add_axes([0, 0, 1, 1])
axes4 = fig2.add_axes([0.2, 0.5, 0.4, 0.4])
```

```python
fig2 = plt.figure()

axes3 = fig2.add_axes([0,0,1,1])
axes4 = fig2.add_axes([0.2, 0.5, 0.4, 0.4])

axes3.plot(x, z)
axes4.plot(x, y)

axes3.set_xlim([x.min(), x.max()])
axes3.set_ylim([z.min(), z.max()])
axes4.set_xlim([20, 22])
axes4.set_ylim([30, 50])

axes3.set_xlabel('X')
axes3.set_ylabel('Z')
axes4.set_xlabel('X')
axes4.set_ylabel('Y')
axes4.set_title('zoom')
```
